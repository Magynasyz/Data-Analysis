# NYC Tree Census and Income: Data Setup

This is Part I of the NYC Tree Census and Income project. By the end of this project, we will have...
- loaded multiple datasets into Tableau
- renamed and formatted columns for the best readability
- joined datasets on a common field
- saved our work privately to Tableau Public
- become familiar with the data and started looking towards making visualizations

You've probably already downloaded all the data you'll need for this project as you worked on the Data Setup lesson. If not, or if you'd like more info on our data, check out the [Github Repository](https://github.com/Codecademy-Curriculum/Learn-Tableau-for-Data-Viz) for this course.

In this project, we'll use the NYC Tree Census, NYC Native Trees dataset, and NYC Income data.
